## Unindo Tabelas

```python
dados_area_urbana = pd.read_csv('dados/area_urbana.csv')
```

```python
dados_area_urbana.info()
```

```python
dados_area_urbana.head(10)
```

## Mantendo todos os dados

```python
# primeiro método
# unindo os dois dataFrames: dados_populacao  e dados_area_urbana
dados_populacao_area_urbana = pd.merge(
    left=dados_populacao, right=dados_area_urbana,
    how='outer', on=['municipio', 'uf'],
    suffixes=[' populacao', ' area urbana']
)
```

```python
dados_populacao_area_urbana.head(10)
```

```python
dados_populacao_area_urbana.tail()
```

```python
dados_populacao_area_urbana.info()
```

```python
dados_populacao_area_urbana.isna().sum()
```

## Mantendo apenas dados existentes nas dias

```python
# segundo método
# unindo os dois dataFrames: dados_populacao  e dados_area_urbana
dados_densidade = pd.merge(
    left=dados_populacao, right=dados_area_urbana,
    how='inner', on=['municipio', 'uf'],
    suffixes=[' populacao', ' area urbana']
)
```

```python
dados_densidade.info()
```

```python
dados_densidade.head()
```

```python
dados_densidade.tail()
```

## Salvando o novo DataFrame

```python
dados_densidade['habitantes por km²'] = dados_densidade['populacao'] / dados_densidade['area urbana km²']
dados_densidade
```

```python
dados_densidade = dados_densidade.sort_values(by='habitantes por km²', ascending=False).reset_index(drop=True)
dados_densidade
```

```python
dados_densidade['ranking densidade'] = dados_densidade.index + 1
dados_densidade
```

```python
dados_densidade = dados_densidade[[
    'municipio', 'uf',
    'populacao', 'ranking populacao',
    'area urbana km²', 'ranking area urbana',
    'habitantes por km²', 'ranking densidade',
]]

dados_densidade
```

```python
# salvando o arquivo
dados_densidade.to_csv('dados/densidade_populacao.csv', index=False)
```
